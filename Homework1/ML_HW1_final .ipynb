```python
#some data processing 
import numpy as np
import csv 
from array import *
import matplotlib.pyplot as plt
from pylab import *
from random import shuffle 
matrix = []
f = False
with open('breast-cancer_cleaned.csv') as DataFile:
    csvReader = csv.reader(DataFile)
    for row in csvReader:
        rows =[]
        rows.append(row[0])
        rows.append(row[1])
        rows.append(row[2])
        if f :
            matrix.append(rows)
        f = True; 
    for row in matrix: 
        if row[0] == "recurrence-events":
            row[0] =1
        else:
            row[0] =-1
    age1 = []
    age2 =[]
    TumorSize1 =[]
    TumorSize2 =[]
    for row in matrix:
        if row[0] == 1:
            age1.append(row[1])
            TumorSize1.append(row[2])
        else:
            age2.append(row[1])
            TumorSize2.append(row[2])
    age =list(set (age1 + age2)) 
    age.sort(key = lambda x : int(x[0:2]))
    TumorSize = list(set(TumorSize1 + TumorSize2))
    TumorSize.sort(key = lambda x : int(x[0:2]) if x[0:2].isdigit() else int(x[0]))
    age_dic ={}
    TumorSize_dic={}
    i = 0
    for element in age:
        age_dic[element] = i
        i+= 1
    i = 0
    for element in TumorSize:
        TumorSize_dic[element] = i
        i+= 1
```

### problem 2.b

```python
#  This perceptron algorithm will start the entire iteration all over again right after it updates the weight. 
class Perceptron:
    def __init__(self,x,y):
        self.pre_weight =[];
        self.iterations = 1;
        self.x = x 
        self.y = y
        self.weight = np.zeros(len(x[1]))
    ## helper function: check previous w for convergency test 
    def check_subarray (self):
        if not self.pre_weight:
            return False
        for item in self.pre_weight:
            if list(item) == list(self.weight):
                return True
            else:
                continue
        return False

    def run_perceptron(self):
        flag = True; 
        while True:
    #         print("in while")
            for i, x_i in enumerate(self.x):
    #             print("in first for")
                if self.y[i] *(np.dot(self.weight, x_i)) <=0:  # update the weight 
                    self.weight = self.weight + self.y[i]*x_i
                    self.iterations += 1
                    if self.check_subarray() and flag: # check if the intercept is necessary
                        self.weight =np.zeros(len(self.x[1]) + 1)
                        b = np.ones((self.x.shape[0], 1))
                        self.x = np.concatenate((self.x,b), axis = 1)
                        flag = False
                        self.pre_weight.clear() 
                        self.iterations = 1;
                        break
                    elif self.check_subarray()and flag == False: #check if it converges with intercept
                        return "This data set is not linearly separable"
                    elif self.check_subarray() == False: 
                        self.pre_weight.append(self.weight)
                    break                   # start from the x_1 again once the weight updated 
                if i == len(self.y) - 1: 
                    return self.weight
```

### problem 2.e

```python
class Perceptron_accuracy(Perceptron):  ##
    def __init__(self,x,y):
        self.training_amount = round(0.7*len(x))
        self.x_train =x[0:self.training_amount]
        self.y_train =y[0:self.training_amount]
        self.x_test = x[self.training_amount:]
        self.y_test = y[self.training_amount:]
        Perceptron.__init__(self,self.x_train,self.y_train)
        self.accuracy_list = []
    def calculate_acu(self,x,y):  ## calculate accuracy 
        self.accuracy_list.clear()
        for w_i in self.pre_weight:
            count = 0;
            for i in range (len(x)):
                if y[i]*(np.dot(w_i[0:2],x[i])+w_i[2]) > 0:
                    count += 1
            accuracy = count / len(x)
            self.accuracy_list.append(accuracy)

    def plot_time_accuracy(self,keyword): ## fucntion for plotting the time vs accuracy
        if keyword == "training":
            self.calculate_acu(self.x_train,self.y_train)
        elif keyword =="test":
            self.calculate_acu(self.x_test,self.y_test)
        time =np.arange(1,self.iterations)
        if keyword =="training":
            plt.plot(time,self.accuracy_list,label="time vs. accuracy for training")
        else:
            plt.plot(time,self.accuracy_list,label="time vs. accuracy for testing")
        plt.xlabel("time")
        plt.ylabel("accuracy")
        plt.legend()
    def plot_training_error_time(self): ## function for plotting the change of error vs accuracy 
        self.calculate_acu(self.x_train,self.y_train)
        error_list = 1 - np.asarray(self.accuracy_list)
        change_of_error =[]
        for i in range(len(error_list) -1 ):
            change_of_error.append(error_list[i+1] - error_list[i])
        error_time = np.arange(2,self.iterations)
        plt.plot(error_time,change_of_error)
        plt.xlabel("time")
        plt.ylabel("empirical error")
        
```

```python
#before shuffled
unshuffled_x = x1
unshuffled_y = y1
unshuffled_accuracy = Perceptron_accuracy(unshuffled_x,unshuffled_y) 
unshuffled_accuracy.run_perceptron()
unshuffled_accuracy.plot_time_accuracy("training")
unshuffled_accuracy.plot_time_accuracy("test")
```

<br>
$\ \ \ \ \ \   $The change of training error for the unshuffled data set is smaller than the training error for the shuffled data in the begining because for the unshuffled data, it keeps training the model for just one label in the begining for a certain amonut time of iterations but the shuffled data trained the model with both labels so it has the higher chance to cause the big fluctutaion from previous training error to now-trained error. After they trained same amount data for both labels (in the later phase), the change of training error becomes more similar which has a constant rate of $\pm 0.2$, however the shuffled data takes only 1/3 time to achieve that.
