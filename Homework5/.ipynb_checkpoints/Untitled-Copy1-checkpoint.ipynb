```python
from scipy.io import loadmat
All_data = loadmat('mnist_all.mat')
train0  =All_data['train0']
test0 = All_data['test0']
train1 = All_data['train1']
test1 = All_data['test0']

import numpy as np 
class NN:
    def __init__(self):
        self.W1= 0.01*np.random.randn(4,784)
        self.W2= 0.01*np.random.randn(4,4)
        self.w3= 0.01*np.random.randn(4)
        self.b1 = 0.01*np.random.randn(4)
        self.b2 = 0.01*np.random.randn(4)
        self.b3 = 0.01*np.random.randn()
    def activation(self,z):
        return 1/(1+np.exp(-z))
    def forward(self,x):
        W = [self.W1,self.W2,self.w3]
        b = [self.b1,self.b2,self.b3]
        h = x
        All_h = []
        for i in range(3):
            h =  W[i]@h + b[i] 
            if i !=2:
                for k, p in enumerate(h):
                    h[k]= self.activation(p)
            else:
                h =self.activation(h)
            All_h.append(h)
        return All_h
    def loss(self,X):
        loss = 0
        for i in X:
            training_features = i[:-1]
            y = i[-1]
            predict = self.forward(training_features)[2]
            loss = loss - (y*np.log(predict) + (1-y)*np.log(1-predict))
        print("loss is ", loss)
        return loss
    def train(self,X):
            i= 0
            loss = 1000
            l = 0
            while loss > 100:
                print("train iteration:",l)
                l = l +1
                h = i + 200
                if h > X.shape[0] - 1:
                    mini_batch = X[i:]
                    r = h-train0.shape[0]
                    mini_batch =np.vstack((mini_batch,X[0:r]))
                    i = h-X.shape[0]
                else:
                    mini_batch = X[i:h]
                    i = h
#                 print("mini_batch", mini_batch.shape)
                D_w3  = np.zeros_like(self.w3)
                D_W2  = np.zeros_like (self.W2)
                D_W1  = np.zeros_like (self.W1)
                D_b3  = np.zeros_like (self.b3)
                D_b2  = np.zeros_like (self.b2)
                D_b1  = np.zeros_like (self.b1)
                loss = self.loss(X)
                for training_sample in mini_batch:
                    layer_output = self.forward(training_sample[:-1])
                    y = training_sample[-1]
                    gradient_z3 = -(y/layer_output[2] +(y-1)/(1-layer_output[2]))*layer_output[2]*(1-layer_output[2])
                    gradient_z2 = gradient_z3 * self.w3 * layer_output[1] * (1-layer_output[1])   
                    gradient_z1 = gradient_z2@self.W2 * layer_output[0] * (1-layer_output[0])
                    d_w3 = gradient_z3 * layer_output[1]
                    d_b3 = gradient_z3 * 1
                    d_W2 = np.array([layer_output[0],]*4)
                    for j in range(gradient_z2.shape[0]):
                                    d_W2[j]  = gradient_z2[j] * d_W2[j] 
                    d_b2 = gradient_z2 *1
                    d_W1 = np.array([training_sample[:-1],]*4)
                    for k in range(gradient_z1.shape[0]):
                                    d_W1[k]  = gradient_z1[k]* d_W1[k] 
                    d_b1 = gradient_z1*1
                    D_w3 = D_w3 + d_w3
                    D_b3 = D_b3 + d_b3            
                    D_W2 = D_W2 + d_W2
                    D_b2 = D_b2 + d_b2
                    D_W1 = D_W1 + d_W1          
                    D_b1 = D_b1 + d_b1
                self.W1= self.W1 - 0.1* D_W1/100
                self.W2= self.W2 - 0.1* D_W2/100
                self.w3= self.w3 - 0.1* D_w3/100
                self.b1 = self.b1 - 0.1 *D_b1/100
                self.b2 = self.b2 - 0.1 *D_b2/100
                self.b3 = self.b3 - 0.1 *D_b3/100      
    def prediction(self,X):
        y = []
        for x in X:
            all_h = self.forward(x)
        if all_h[2] > 0.5:
            y.append(1)
        else:
            y.append(0)
        return y 
```
